## Model Creation - Now let us define the Decision Tree Algorithm

## Now visualizing the Decision Tree to understand it better

```python
from sklearn import tree
clf=tree.DecisionTreeClassifier()
clf=clf.fit(X,y)
```

```python
fig = plt.figure(figsize=(20,20))
plot_tree = tree.plot_tree(clf,feature_names = df.columns, class_names = df['Species'].unique().tolist(),precision=4,label = 'all',
                          filled = True)
```
